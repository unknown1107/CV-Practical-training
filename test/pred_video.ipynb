```python
import os
import cv2
from ultralytics import YOLO

# parameters
MODEL_PATH = "/home/huangliang/CV-Practical-training/runs/detect/train2/weights/best.pt"
VIDEO_ROOT_PATH = "./videos"
ORIGIN_VIDEO_PATH = "1.mp4"
OUTPUT_VIDEO_PATH = "1_only_helmet.mp4"

# 补全视频路径
ORIGIN_VIDEO_PATH = os.path.join(VIDEO_ROOT_PATH, 'origin', ORIGIN_VIDEO_PATH)
OUTPUT_VIDEO_PATH = os.path.join(VIDEO_ROOT_PATH, 'pred', OUTPUT_VIDEO_PATH)

# 加载YOLOv8模型
model = YOLO(MODEL_PATH)  # 替换为你自己的模型路径


# 打开视频文件
cap = cv2.VideoCapture(ORIGIN_VIDEO_PATH)

# 获取视频帧的维度
frame_width = int(cap.get(3))
frame_height = int(cap.get(4))

#创建VideoWriter对象
fourcc = cv2.VideoWriter_fourcc(*'mp4v')
out = cv2.VideoWriter(OUTPUT_VIDEO_PATH, fourcc, 20.0, (frame_width, frame_height))

#循环视频帧
while cap.isOpened():
    # 读取某一帧
    success, frame = cap.read()
    if success:
        # 使用yolov8进行预测
        results = model(frame, classes=[1, 2])
        #可视化结果
        annotated_frame = results[0].plot()
        #将带注释的帧写入视频文件
        out.write(annotated_frame)
    else:
        # 最后结尾中断视频帧循环
        break
#释放读取和写入对象
cap.release()
out.release()
```
